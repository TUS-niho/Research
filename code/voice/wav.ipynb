```python
import opensmile
import audiofile
import numpy as np
import pandas as pd # pandasをインポート

# --- 1. 準備 ---
VIDEO_FPS = 29.97
wav_path = "./ID36_audio.wav"

try:
    signal, sampling_rate = audiofile.read(wav_path)
except FileNotFoundError:
    print(f"エラー: ファイル '{wav_path}' が見つかりません。")
    exit()

if signal.ndim == 2:
    signal = signal[0, :]


# --- 2. スライス ---
start_video_frame = 156
end_video_frame = 222

start_sec = start_video_frame / VIDEO_FPS
end_sec = end_video_frame / VIDEO_FPS
start_sample = int(start_sec * sampling_rate)
end_sample = int(end_sec * sampling_rate)

signal_slice = signal[start_sample:end_sample]


# --- 3. 分析 & 時間情報の追加 ---
if signal_slice.size > 0:
    smile = opensmile.Smile(
        feature_set=opensmile.FeatureSet.eGeMAPSv02,
        feature_level=opensmile.FeatureLevel.LowLevelDescriptors,
    )
    result_df = smile.process_signal(
        signal_slice,
        sampling_rate
    )

    # --- ▼ ここからが追加部分 ▼ ---
    
    # 1. スライス開始時間をTimedelta型に変換
    start_timedelta = pd.to_timedelta(start_sec, unit='s')

    # 2. 元のインデックスに開始時間を加算して、新しい列を作成
    #    MultiIndexの'start'と'end'を取得して計算
    original_start = result_df.index.get_level_values('start') + start_timedelta
    original_end = result_df.index.get_level_values('end') + start_timedelta

    # 3. 新しい列をDataFrameに追加
    result_df['original_start'] = original_start
    result_df['original_end'] = original_end
    
    # --- ▲ ここまで ▲ ---

    # 'answer1'という名前の列を0番目（先頭）に挿入し、値を1に設定
    result_df.insert(0, 'label', 'answer1')
    
    print(f"✅ 元の時間情報を追加した分析結果:")
    print(result_df.head()) # 新しい列が追加されたことを確認

    # CSVファイルに保存
    output_filename = "analysis_results_with_original_time.csv"
    result_df.to_csv(output_filename)
    print(f"\n✅ 分析結果を '{output_filename}' に保存しました。")

else:
    print("⚠️ エラー: スライス後のデータが空です。")
```

```python
import opensmile
import pandas as pd

# --- 1. カスタム設定ファイルを指定してインスタンスを作成 ---
smile = opensmile.Smile(
    # ↓ この2行は必須だが、設定ファイルで上書きされるため無視される
        feature_set=opensmile.FeatureSet.eGeMAPSv02,
        feature_level=opensmile.FeatureLevel.LowLevelDescriptors,
    # ↑ ----------------------------------------------------
    options={
        '-C': 'custom_window_config.conf'  # 作成した設定ファイルを指定
    }
)

# --- 2. 音声ファイルを処理 ---
# ▼▼▼ デモ音声の代わりに、ご自身のファイルパスを指定 ▼▼▼
audio_file = "./ID36_audio.wav"

try:
    result_df = smile.process_file(audio_file)

    # --- 3. 結果の確認 ---
    print(f"✅ カスタム設定ファイルを使って特徴量を抽出しました。")

    # タイムスタンプの差から、フレームステップが0.010秒になっていることを確認
    if len(result_df) > 1:
        frame_step = result_df.index[1][1] - result_df.index[0][1]
        print(f"フレームステップ: {frame_step.total_seconds()} 秒")
    
    print("\n分析結果:")
    print(result_df.head())

except FileNotFoundError:
    print(f"エラー: ファイル '{audio_file}' が見つかりません。パスを確認してください。")
except Exception as e:
    print(f"エラーが発生しました: {e}")
```

```python
import opensmile
import pandas as pd

# 1. MFCCの特徴量セットを選択
#    同時に、optionsでフレームサイズとステップを10ms（0.010秒）に上書き
smile = opensmile.Smile(
    feature_set=opensmile.FeatureSet.MFCC12_0_D_A,
    feature_level=opensmile.FeatureLevel.LowLevelDescriptors,
    options={
        'frameSize': 0.010,
        'frameStep': 0.010,
    }
)

# 2. 音声ファイルを処理
audio_file = "./ID36_audio.wav"  # ご自身のファイルパスを指定してください

try:
    result_df = smile.process_file(audio_file)

    # 3. 結果の確認
    print("✅ 10msのフレームサイズでMFCCを抽出しました。")
    
    # タイムスタンプの差から、フレームステップが0.010秒になっていることを確認
    if len(result_df) > 1:
        start_time = result_df.index[0][1]
        end_time = result_df.index[0][2]
        frame_duration = end_time - start_time
        
        print(f"フレームの長さ: {frame_duration.total_seconds()} 秒") # 0.010になるはず
    
    print("\n分析結果:")
    print(result_df.head())

except AttributeError:
    print("エラー: お使いのopensmileのバージョンでは 'MFCC12_0_D_A' が利用できないようです。")
    print("以前の方法（カスタム設定ファイル）をお試しください。")
except Exception as e:
    print(f"エラーが発生しました: {e}")
```

```python
import os
import subprocess
import pandas as pd
import soundfile

# --- 1. 設定と準備 ---
config_str = """
frameSize = 0.010
frameStep = 0.010
[componentInstances:cComponentManager]
instance[energy].type = cEnergy
instance[spectral].type = cSpectral
instance[mfcc].type = cMfcc
instance[lld_concat].type = cVectorConcat
instance[csvSink].type = cCsvSink
[energy:cEnergy]
reader.dmLevel = pcm
writer.dmLevel = energy_lld
rms = 1; log = 1; zcr = 1
[spectral:cSpectral]
reader.dmLevel = pcm
writer.dmLevel = spectral_lld
[mfcc:cMfcc]
reader.dmLevel = pcm
writer.dmLevel = mfcc_lld
[lld_concat:cVectorConcat]
reader.dmLevel = energy_lld;spectral_lld;mfcc_lld
writer.dmLevel = lld_all
[csvSink:cCsvSink]
reader.dmLevel = lld_all
filename = \\cm[outputfile(I.csv)]
delimChar = ;
printHeader = 1
"""
config_filename = "10ms_custom_lld_config.conf"
with open(config_filename, "w") as f:
    f.write(config_str)

# --- 2. 実行の準備 ---
wav_path = "./ID36_audio.wav"
temp_csv_path = "temp_output.csv"

# ▼▼▼ ここが最重要ポイント ▼▼▼
# SMILExtract.exeの完全なパスを直接指定する
smilextract_path = r"C:\Users\akini\anaconda3\envs\audio_analysis_env\Lib\site-packages\opensmile\core\bin\win_amd64\SMILExtract.exe"

# 実行するコマンドを作成
command = [
    smilextract_path,  # ← 直接指定したパスを使う
    '-C', config_filename,
    '-I', wav_path,
    '-O', temp_csv_path,
]

# --- 3. SMILExtractを直接実行 ---
print(f"SMILExtractを直接実行して '{wav_path}' を分析します...")
try:
    subprocess.run(command, check=True, capture_output=True, text=True)

    result_df = pd.read_csv(temp_csv_path, sep=';')
    print("✅ 10msフレームでの特徴量抽出に成功しました。")
    print(result_df.head())

except FileNotFoundError:
    print(f"エラー: 指定されたパスに 'SMILExtract.exe' が見つかりません。")
    print(f"パスを確認してください: {smilextract_path}")
except subprocess.CalledProcessError as e:
    print("SMILExtractの実行中にエラーが発生しました。")
    print("--- エラー出力 ---\n", e.stderr)
finally:
    # --- 4. 一時ファイルを削除 ---
    if os.path.exists(config_filename): os.remove(config_filename)
    if os.path.exists(temp_csv_path): os.remove(temp_csv_path)
```
