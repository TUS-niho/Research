```python
import os
import pandas as pd
import glob

folder_path = './processed_annotation'  # フォルダーのパスを指定

# 質問範囲が記載されたCSVファイルのリストを取得（*_annotation_processed.csvに一致するファイル）
annotation_files = glob.glob(os.path.join(folder_path, '*_annotation_processed.csv'))

# 結合するための空のリストを作成
all_combined_data = {}

# 各質問ファイルに対する処理
for annotation_file in annotation_files:
    # 対応するデータファイル名を生成
    base_name = os.path.basename(annotation_file).replace('_annotation_processed.csv', '.csv')
    data_file_path = os.path.join(folder_path, base_name)
    
    # データファイルが存在するか確認
    if not os.path.exists(data_file_path):
        print(f"Data file for {base_name} not found, skipping...")
        continue        

    # 質問範囲ファイルとデータファイルの読み込み
    annotation_df = pd.read_csv(annotation_file, header=None, names=['target', 'start_time', 'end_time', 'duration_time', 'number', 'start_frame', 'end_frame', 'duration_time_2'], skipinitialspace=True)
    data_df = pd.read_csv(data_file_path)
    
    # 各範囲ごとの処理
    for _, row in annotation_df.iterrows():
        start, end = row['start_frame'], row['end_frame']
        number = row['number']
        
        # 指定範囲のデータを抽出
        data_df['frame'] = data_df.index  # 'frame'がすでにある場合はその列名に変更
        annotation_data = data_df[(data_df['frame'] >= start) & (data_df['frame'] <= end)].copy()
        
        # IDごとにデータを結合
        if number not in all_combined_data:
            all_combined_data[number] = annotation_data
        else:
            all_combined_data[number] = pd.concat([all_combined_data[number], annotation_data], ignore_index=True)
        
        # 結果の確認: 抽出したデータの最初の数行を出力
        print(f"Extracted data for ID {number} from {base_name}:")
        print(annotation_data.head())  # 抽出したデータの最初の数行を表示
        print('-' * 40)  # 区切り線

# 保存先のフォルダを作成
output_folder = './extracted_data'
os.makedirs(output_folder, exist_ok=True)

# 各IDごとに結合したデータをCSVファイルとして保存
for number, combined_data in all_combined_data.items():
    output_file = os.path.join(output_folder, f"ID{number}_combined_extracted.csv")
    combined_data.to_csv(output_file, index=False)
    print(f"Saved {output_file}")
```

```python
import pandas as pd

# CSVファイルを読み込む
df1 = pd.read_csv('./processed_annotation/ID30.csv')
df2 = pd.read_csv('./processed_annotation/ID30_annotation_processed.csv', header=None, usecols=[5, 6, 7])

# 結果を格納するためのDataFrameを初期化
result = pd.DataFrame()

# 各行の範囲に基づいて対応する行を df1 から抽出
for _, row in df2.iterrows():
    start_index = int(row[6])  # 開始インデックス（整数に変換）
    end_index = int(row[7])    # 終了インデックス（整数に変換）

    # インデックスに対応する行を抽出
    extracted = df1.iloc[start_index:end_index + 1]  # ilocで範囲抽出（終了インデックスを含む）

    # "5"列の値を新しい列として追加（question名を記録）
    extracted['label'] = row[5]
    
    # ラベル列を先頭に移動
    extracted.insert(0, 'label', extracted.pop('label'))

    # 抽出したデータを結合
    result = pd.concat([result, extracted], ignore_index=True)

# 抽出結果を確認
#print(result)

# 必要ならCSVに保存
result.to_csv('./output_filtered_questions.csv', index=False)
```
